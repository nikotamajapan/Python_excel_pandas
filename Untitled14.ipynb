```python
def extract(ff):
  file = glob.glob('/content/*.xlsx')
  ff=file[0]
  _ff=pd.read_excel(ff)
  columns = _ff.iloc[10,[1,2,4,10,11,14]]
  ff=_ff.iloc[11:23 ,[1,2,4,10,11,14]]
  ff.columns = columns
  ff['company']=_ff.iloc[2,0]
  ff['com code']=_ff.iloc[3,4]
  ff['inv no']=_ff.iloc[2,12]
  ff['date']=_ff.iloc[3,12]
  ff['issue']=_ff.iloc[4,12]
  ff['issue code']=_ff.iloc[4,13]
  return ff
  
```
